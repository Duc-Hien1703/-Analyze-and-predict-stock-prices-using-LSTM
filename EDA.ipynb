```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('RTSI.csv')
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
for column in df.columns:
    plt.figure(figsize=(12,6))
    plt.title(f'Distribution of {column}')
    plt.hist(df[column], bins=30)
    plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Open and Close Prices Over Time')
plt.plot(df['Open'], label='Open')
plt.plot(df['Close'], label='Close')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Volatility Prices Over Time')
plt.plot(df['Volatility'], label='Volatility')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Open and Close Prices Over Time')
plt.plot(df['Open'].tail(300), label='Open')
plt.plot(df['Close'].tail(300), label='Close')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Volatility Prices Over Time')
plt.plot(df['Volatility'].tail(300), label='Volatility')
plt.legend()
plt.show()
```

```python
df['Daily Return'] = df['Close'].pct_change()
df['Daily Return'].plot(figsize=(12,6), legend=True, linestyle='--', marker='o')
```

```python
df['Daily Return'] = df['Open'].pct_change()
df['Daily Return'].plot(figsize=(12,6), legend=True, linestyle='--', marker='o')
```
